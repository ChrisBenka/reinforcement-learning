```python
import gym
env = gym.make('CartPole-v0')
env.reset()
for _ in range(1000):
    env.render()
    env.step(env.action_space.sample()) # take a random action
env.close()
```

## observations ##
- observation: (object) represents your observation of the env e.g. pixel data from a camera, joint angles + joint velocities of a robot, or borad state in a board game
- reward (float): amount of reward achieved by previous action
- done whether its time to reset the env (indicates episode has terminated)
- info diagnostic info useful for debugging sometimes useful for learning (might contain raw probs behind env's last state change)



```python
import gym 
env = gym.make('CartPole-v0')
for i_episode in range(20):
    observation = env.reset()
    for t in range(100):
        env.render()
        print(observation)
        action = env.action_space.sample()
        observation,reward,done,info = env.step(action)
        if done:
            print("Episode finished after {} timesteps".format(t+1))
            break
env.close()
```
